Let $Θ = \{θ_1,...,θ_k\}$ be a finite parameter space. Prove that the posterior mode is the Bayes estimator under zero–one loss.


The Bayes rule minimizes the Bayes Risk:
$r( \hat{θ} | x ) = \int L(θ, \hat{θ})f(θ|x)dθ = \sum I(\hat{θ} \neq θ)f(θ|x) = 1 - \sum I(\hat{θ} = θ)f(θ|x) = 1 - f(\hat{θ} | x )$

Minimising the 0-1 Loss is equivalent to maximasing the Posterior mode (posterior max). A similar derivation can be done with continuous random variables, using the dirac delta function in place of the indicator function

```python
def mle(X):
  return X

def jse(X):
  return np.maximum(1 - ((len(X) - 2) / sum(X**2)), 0) * X

def mse(theta_pred, theta):
    return ((theta - theta_pred)**2).mean()
```

```python
sizes = [3, 30, 300, 3000]
B = 10000
for n in sizes:
  error_mle = np.empty(B)
  error_jse = np.empty(B)
  for i in range(B):
    theta = np.ones(n)
    X = norm.rvs(loc=1, scale=1, size=n )
    theta_hat_mle = mle(X)
    theta_hat_jse = jse(X)
    error_mle[i] = mse(theta_hat_mle, theta)
    error_jse[i] = mse(theta_hat_jse, theta)

  plt.figure(figsize=(12, 8))
  plt.hist(error_mle, density=True, bins=100, histtype='step', color='blue', label='MLE, n = ' + str(n))
  plt.hist(error_jse, density=True, bins=100, histtype='step', color='red', label='James-Stein, n = ' + str(n))
  plt.legend(loc='upper right')
  plt.show()


```
